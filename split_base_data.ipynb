```python
import pandas as pd # Read/write data
from sklearn.model_selection import train_test_split # Train-test split of dataset

base_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ECE697/Project/Base.csv")
```

```python
base_train_df.to_csv("/content/drive/MyDrive/Colab Notebooks/ECE697/Project/base_train.csv", encoding='utf-8') # writing training df to csv
base_val_df.to_csv("/content/drive/MyDrive/Colab Notebooks/ECE697/Project/base_val.csv", encoding='utf-8') # writing testing df to csv
```
