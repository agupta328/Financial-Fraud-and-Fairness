```python
import matplotlib.pyplot as plt
import numpy as np

# creation of income bins between 0.1-0.9 to serve as x-axis and understand distribution of fraud samples
income_bins=[]
for i in range(1, 10):
  income_bins.append(i / 10)
centers = income_bins[:-1] + np.diff(income_bins) / 2 # calculates center of income_bins and adds 0.5 for next value

frequency_fraud_incomes = [freq_income_01_fraud, freq_income_02_fraud, freq_income_03_fraud, freq_income_04_fraud, freq_income_05_fraud, freq_income_06_fraud, freq_income_07_fraud, freq_income_08_fraud]
fraud_incomes = [fraud_income_01, fraud_income_02, fraud_income_03, fraud_income_04, fraud_income_05, fraud_income_06, fraud_income_07, fraud_income_08]
non_fraud_incomes = [non_fraud_income_01, non_fraud_income_02, non_fraud_income_03, non_fraud_income_04, non_fraud_income_05, non_fraud_income_06, non_fraud_income_07, non_fraud_income_08]

# creating three subplots to display large dataset distributions
fig, axs = plt.subplots(3, 1, figsize=(7, 15))

# plot fraud frequency vs income bins
axs[0].stairs(frequency_fraud_incomes, edges=income_bins, edgecolor='black', label="Fraud frequency")
axs[0].plot(centers, frequency_fraud_incomes, 'o-', color='blue', alpha=0.3)
axs[0].set_xlabel("Income levels")

# plot fraud sample counts vs income bins
axs[1].stairs(fraud_incomes, edges=income_bins, edgecolor='black', label="Fraud counts")
axs[1].plot(centers, fraud_incomes, 'o-', color='blue', alpha=0.3)
axs[1].set_xlabel("Income levels")

# plot non-fraud sample counts vs income bins
axs[2].stairs(non_fraud_incomes, edges=income_bins, edgecolor='black', label="Non-fraud counts")
axs[2].plot(centers, non_fraud_incomes, 'o-', color='blue', alpha=0.3)
axs[2].set_xlabel("Income levels")

for ax in axs:
    ax.legend()

plt.show()
```
